# groupby 함수 : 범주 묶기
groupby 함수를 이용해서 데이터프레임의 데이터들을 범주별로 묶어서 통계를 내거나 출력할 수 있습니다.

```python
import pandas as pd
import numpy as np
# 데이터프레임 인덱스 성분, 칼럼 성분 정의
columns = ['시','구', '동', '휘발유 가격','경유 가격']
data = [['서울','노원구','상계1동', 1997, 1738],
        ['서울','노원구','월계동', 2073, 2010],
        ['서울','강남구','개포동', 2103, 2073],
        ['서울','강남구','도곡동', 2057, 2023],
        ['서울','서대문구','신촌동', 2021, 1999],
        ['서울','서대문구','북가좌동', 2025, 2000],
        ['부산','강서구','대저1동', 1988, 1956],
        ['부산','강서구','강동동', 1972, 1937],
        ['부산','해운대구','우동', 2019, 1999],
        ['부산','해운대구','석대동', 2001, 1981],
        ['부산','동래구','온천동', 1999, 1941],
        ['부산','동래구','낙민동', 1986, 1954]]
```

```python
# 데이터 프레임 생성
df = pd.DataFrame(data,columns=columns)
df
```

```python
# 범주로 쓸 칼럼의 성분으로 범주를 나눔 ('시')
groups = df.groupby("시")
groups.size()
```

```python
# 시별 평균값 출력
groups.mean()
```

```python
# 시별 최대값 출력
groups.max()
```

```python
# 지역구별 min, max값 출력
groups.aggregate([min,max])
```

```python
# 지역구별 기름값 평균, 표준편차 출력
groups.aggregate(['mean','std'])
```

## groupby로 두개 이상 칼럼으로 범주를 나누기
groupby는 두개 이상의 칼럼을 기준으로 그룹화도 가능합니다.

```python
group_by_two = df.groupby(['시','구'])
group_by_two.size()
```

```python
# 지역구별 기름값 평균확인
group_by_two.mean()
```
